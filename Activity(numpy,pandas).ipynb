```python
import pandas as pd
import numpy as np

df = pd.read_csv("titanic_dataset.csv")
```

```python
print(df.head(10))
```

```python
print(df.isnull().sum())
```

```python
print(df.groupby("Sex")["Survived"].mean() * 100)
```

```python
print("Average:", df["Age"].mean())
print("Median:", df["Age"].median())
print("Std:", df["Age"].std())
```

```python
print(df["Pclass"].value_counts())
```

```python
print(df[(df["Sex"]=="female") &
         (df["Age"]<18) &
         (df["Survived"]==1)])
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
```

```python
def age_group(age):
    if age < 18:
        return "Child"
    elif age <= 60:
        return "Adult"
    else:
        return "Senior"

df["age_group"] = df["Age"].apply(age_group)
```

```python
print(df.nlargest(5, "Fare"))
```

```python
print(df.groupby(["Pclass","Sex"])["Fare"].mean())
```

```python
df.rename(columns={
    "Sex":"gender",
    "Fare":"ticket_fare"
}, inplace=True)
```

```python
print(df.groupby(["Pclass","Survived"]).size())
```
