```python
import pandas as pd
import os
import time
from datetime import datetime
import numpy as np
import requests
import sqlite3
```

**Data Extraction**

```python
#task1
population = pd.read_csv("/content/population.csv")
print(population.head(5))
print(population.shape)
print(population.dtypes)
```

```python
#task2
gdp=pd.read_excel("/content/gdp.xlsx")
print(gdp.columns)
print(gdp.dtypes)
```

```python
#task3
in_users=pd.read_json("/content/internet_users.json")
users=pd.DataFrame(in_users)
users.head()
```

```python
#task 5
literacy=pd.read_xml("/content/literacy_rate.xml")
literacy=pd.DataFrame(literacy)
literacy.head()
```

```python
#task 5
#population
print(f"missing values: {population.isnull().sum()}")
print(f"duplicate values: {population.duplicated().sum()}")

#gdp
print(f"missing values: {gdp.isnull().sum()}")
print(f"duplicate values: {gdp.duplicated().sum()}")

#internet users
print(f"missing values: {users.isnull().sum()}")
print(f"duplicate values: {users.duplicated().sum()}")

#literacy
print(f"missing values: {literacy.isnull().sum()}")
print(f"duplicate values: {literacy.duplicated().sum()}")
```

**Data Transformation**

```python
#task 6
gdp.rename(columns={'Country':'country'}, inplace=True)
literacy.rename(columns={'name':'country'}, inplace=True)
```

```python
print(population.columns)
print(gdp.columns)
print(users.columns)
print(literacy.columns)
```

```python
print(population.dtypes)
print(gdp.dtypes)
print(users.dtypes)
print(literacy.dtypes)
```

```python
#task 7
#task 8

data_df=population.merge(gdp,on='country')\
.merge(users,on='country')\
.merge(literacy,on='country')

data_df.head()
```

```python
#task 9
data_df['Internet_Penetration(%)']=((data_df['internet_users'] / data_df['population']) * 100).round(2)
data_df['GDP Per Capita']= (data_df['GDP'] / data_df['population']).round(2)

data_df.head()
```

**Load Data**


```python
#task 10

conn = sqlite3.connect('country_data.db')
data_df.to_sql('country_statistics',conn, if_exists='replace',index=False)
cursor=conn.cursor()
```

```python
#task 11
query='''select * from country_statistics limit 5'''
cursor.execute(query)
countries=cursor.fetchall()

for row in countries:
  print(row)
```

**Data Analysis Using SQL**

```python
#11
cursor.execute("select country,GDP from country_statistics order by GDP desc limit 5")

top_gdp_countries = cursor.fetchall()

for row in top_gdp_countries:
    print(row)
```

```python
#12
query='''
select avg(literacy_rate)
from country_statistics'''

cursor.execute(query)
top_gdp_countries = cursor.fetchall()

for row in top_gdp_countries:
    print(row)
```

```python
#13
query='''
select country,literacy_rate
from country_statistics
where literacy_rate > 90
order by literacy_rate desc '''

cursor.execute(query)
top_gdp_countries = cursor.fetchall()

for row in top_gdp_countries:
    print(row)
```

```python
#14
query='''
select country,"Internet_Penetration(%)"
from country_statistics
order by 'Internet_Penetration(%)' desc
limit 5'''

cursor.execute(query)
top_gdp_countries = cursor.fetchall()

for row in top_gdp_countries:
    print(row)
```

```python

query = '''
ALTER TABLE country_statistics
RENAME COLUMN `Internet_Penetration(%)` TO internet_penetration;
'''

cursor.execute(query)
conn.commit()

```

```python
query = '''
ALTER TABLE country_statistics
RENAME COLUMN `GDP Per Capita` TO GDP_per_capita;
'''

cursor.execute(query)
conn.commit()
```
