## Challenge 2 : Analyze daily expenses and extract insights.
User enters:

Number of transactions
Category and amount

Food 200
Travel 500
Food 300
Shopping 700

### Output
* Total expense
* Highest expense category
* Category-wise total

```python
def analyze_expenses():
    n = int(input("Enter number of transactions: "))

    expenses = {}

    for _ in range(n):
        category, amount = input().split()
        amount = float(amount)

        # TODO: store category-wise total
        expenses[category]=expenses.get(category,0)+amount


    total=sum(expenses.values())
    # TODO: find highest category
    highest_category="Empty"
    highest_amount=0
    for category,amount in expenses.items():
      if amount>highest_amount:
        highest_amount=amount
        highest_category=category


    print("Total Expense:", total)
    print("Highest Category:", highest_category)
    print("Category-wise:", expenses)

analyze_expenses()
```

ff### Challenge 3: Data Cleaner (Mini Pandas Logic Without Pandas)

10 ,20 ,abc ,30 ,-5 , 40

* Remove invalid values
* Remove negative values


Show:
* Clean list
* Average
* Max value

```python
def clean_data():
    n = int(input("Enter number of values: "))

    data = []

    for _ in range(n):
        value = input()

        try:
            num = float(value)

            if num<0:
              continue

            data.append(num)

        except ValueError:

            pass

    avg=sum(data)/len(data)
    max_val=max(data)

    print("Clean Data:", data)
    print("Average:", avg)
    print("Max:", max_val)

clean_data()
```
