1. **Bank Account Manager**


```python
class BankAccnt:
  def __init__(self,user,accnt,bal):
    self.user=user
    self.accnt=accnt
    self.bal=bal

  def deposit(self,amount):
    if amount <=0:
      print(f"You have entered {amount}\nPlease enter a valid Amount.")
    else:
      self.bal+=amount
      print(f"Cash deposited successfully.\nCurrent Balance: {self.bal}Rs")

  def withdraw(self,amount):
    if amount>0:
      if amount>self.bal:
        print("Insufficient Balance")
      else:
        self.bal-=amount
        print(f"{amount}Rs withdrawn successfully.\nCurrent Balance: {self.bal}Rs")
    else:
      print("Enter a valid Amount")

  def balance(self):
    print(f"Account Holder: {self.user}\nAccount No: {self.accnt}\nCurrent Balance: {self.bal}Rs")



name=input("Enter the name: ")
accnt=input("Enter the Account Number: ")
obj1=BankAccnt(name,accnt,0)
while True:

  op=int(input("What operation do you want to Perform\n1.Deposit\n2.Withdrawal\n3.Balance Enquiry\n4.Exit\n"))

  if op==1:
    amount=int(input("Enter the Amount to deposit"))
    obj1.deposit(amount)
  elif op==2:
    amount=int(input("Enter the amount to Withdraw"))
    obj1.withdraw(amount)
  elif op==3:
    obj1.balance()
  elif op==4:
    break
  else:
    print("enter a valid Number")









```

**2.Weather Data Analyzer**


```python
temp=[]
print("Enter the Temperature of past 3 Hrs:\n")
temp.append(int(input("1: ")))
temp.append(int(input("2: ")))
temp.append(int(input("3: ")))

humi=[]
print("Enter the Humidity of past 3 Hrs:\n")
humi.append(int(input("1: ")))
humi.append(int(input("2: ")))
humi.append(int(input("3: ")))


def avg(a):
  avg=sum(a)/len(a)
  return avg

avg_temp=avg(temp)
avg_humi=avg(humi)

if avg_humi >50 and avg_temp >30 :

  print("Sunny")
elif avg_humi > 80 and avg_temp < 30:
    print("Rainy")

elif avg_temp > 30 and avg_humi > 70:
    print("Hot & Humid")

else:
    print("Cloudy")



```
