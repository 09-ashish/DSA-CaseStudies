```python
#task7
data['genre_count']=data.Genre.apply(
    lambda x: len(x.split(',')))


data.head(5)[['Title','genre_count']]
```

```python
#task8
data['Revenue (Millions)']=data['Revenue (Millions)'].fillna(data['Revenue (Millions)'].median())
data['Metascore']=data.Metascore.fillna(data.Metascore.median())

print(data[['Revenue (Millions)','Metascore']].isna().sum())
```

```python
#task17
data['Category']=data.Rating.apply(
    lambda x: "Poor" if x<6
    else "Average" if x<7.5
    else "Excellent"
)
data.sample(5)
```

```python
#task18
def movie_score(row):
  return (row['Rating']*10) + row['Revenue (Millions)']

data['High_Score']=data.apply(
    movie_score,axis=1)

data.head(5)[['Title','Rating','Revenue (Millions)','High_Score']]
```
